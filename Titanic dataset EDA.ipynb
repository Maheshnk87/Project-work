```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
###Loading the dataset

df_train=pd.read_csv('C:/Users/hp/Desktop/Datasets/titanic/train.csv')
df_test=pd.read_csv('C:/Users/hp/Desktop/Datasets/titanic/test.csv')
```

```python
### Prepreocessing the data
def pre_processing (df_train):
    rows=df_train.shape[0]
    print(rows)
    column=df_train.shape[1]
    print(column)
    total_miss=df_train.isnull().sum()
    print(total_miss)
    return(df_train)
```

```python
pre_processing(df_train)
```

```python
def bar_plot(variable):
    var=df_train[variable]
    var_value=var.value_counts()
    sns.barplot(x=var_value.index,y=var_value,data=df_train,)
    plt.show()
    print("{}:\n{}".format(variable,var_value))
```

```python
cat1=["Survived","Sex","Pclass","Embarked","SibSp", "Parch"]
```

```python
for c in cat1:
    bar_plot(c)
```

```python
def plot_hist(variable):
    sns.histplot(df_train[variable])
    plt.show()
```

```python
num=["Fare", "Age",]
```

```python
for n in num:
    plot_hist(n)
```
